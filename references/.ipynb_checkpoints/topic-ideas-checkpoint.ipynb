## Analyse von Preisdaten für Lieferkettenlieferungen

### Data source

Quelle: https://data.usaid.gov/HIV-AIDS/Supply-Chain-Shipment-Pricing-Data/a3rc-nmf6


### Data characterisitcs

Der Datensatz bietet umfassende Informationen zu Supply Chain Health Ware, einschließlich Versand- und Preisdaten für Lieferungen von Antiretroviralen (ARV) und HIV-Laborprodukten in unterstützte Länder. Er enthält auch Daten zu Rohstoffpreisen und den damit verbundenen Lieferkettenkosten für den Transport von Rohstoffen in Nutzungsländer.
Der vorliegende Datensatz besteht aus einer Stichprobe von insgesamt 10.324 Datensätzen mit 33 numerische und kategoriale Variablen. Einige Variablen können als eindeutige unabhängige Variablen identifiziert werden. Auch dieser Datensatz erfüllt somit die Projektanforderungen.


### Research question

**Lineare Regression:**

Wie variieren die Kosten und Volumina von Antiretroviralen (ARV) und HIV-Laborprodukten in der Supply Chain Health Ware, und welche Einflussfaktoren, einschließlich Projektcode, Versandart, Land, Lieferant, Produktgruppe, geplantes Lieferdatum usw., tragen zu diesen Variationen bei?

Für das lineare Regressionsmodell würden wir die Kosten oder Volumina von Antiretroviralen (ARV) und HIV-Laborprodukten als abhängige Variable und die folgenden Einflussfaktoren als unabhängige Variablen betrachten:
   - Projektcode (identifiziert spezifische Projekt zu dem die Lieferung gehört.)
   - Versandart (auf welche Weise die Lieferung erfolgt.)
   - Land (Zielland, in dem die ARV- und HIV-Laborprodukte empfangen werden sollen.)
   - Lieferant (Lieferant, von dem die Gesundheitsprodukte bezogen werden.)
   - Produktgruppe (Gruppe, zu der das Gesundheitsprodukt gehört, z.B. ARV oder HIV-Laborprodukte.)
   - Geplantes Lieferdatum (vorhergesehene Datum, an dem die Lieferung erfolgen soll.)
   - Frachtkosten in USD ( Kosten für den Frachttransport der Lieferung.)

**Klassifikation:**

Wie können wir basierend auf verschiedenen Variablen wie Lieferant, Versandart, Land usw. die Produktgruppe (ARV, HIV-Laborprodukte, etc.) vorhersagen?

Abhängige Variable: Produktgruppe (Gruppe, zu der das Gesundheitsprodukt gehört, z.B. ARV oder HIV-Laborprodukte.)

Unabhängige Variablen: Andere relevante Variablen wie Lieferant, Versandart, Land, geplantes Lieferdatum, etc.
